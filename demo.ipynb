```python
def simple_gen():
    print("Start")
    yield 1
    print("Middle")
    yield 2
    print("End")
```

```python
for n in simple_gen():
    print(n)
```

```python
def nums():
    print("A")
    yield 1
    print("B")
    yield 2
    print("C")
    yield 3
    print("D")

g = nums()

next(g)  # prints "A", returns 1
next(g)  # prints "B", returns 2
next(g)  # prints "C", returns 3
next(g)  # prints "D", then StopIteration (khatam)
```
